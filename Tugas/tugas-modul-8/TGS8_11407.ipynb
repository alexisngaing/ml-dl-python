```python
# Ubah data string menjadi data numeric

df_stroke["gender"].replace({"Male": 1, "Female": 0, "Other": -1}, inplace=True)
df_stroke["ever_married"].replace({"Yes": 1, "No": 0}, inplace=True)
df_stroke["work_type"].replace({'Private': 0, "Self-employed": 1, "Govt_job": 2, "children": 3, "Never_worked": 4}, inplace=True)
df_stroke["Residence_type"].replace({"Urban": 1, "Rural": 0}, inplace=True)
df_stroke["smoking_status"].replace({"smokes": 2, "formerly smoked": 1, "never smoked": 0, "Unknown": -1}, inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, KFold, StratifiedKFold

X = df_stroke.drop(['id', 'stroke'], axis=1)
y = df_stroke['stroke']

logreg = LogisticRegression(max_iter=1000)
```

```python
kfold = KFold(n_splits=3, shuffle=False)
skfold = StratifiedKFold(n_splits=3, shuffle=False)

scores3_kfold = cross_val_score(logreg, X.values, y.values, cv=kfold)
print("3-Fold Cross Validation Score: \n{}".format(scores3_kfold))

scores3_skfold = cross_val_score(logreg, X.values, y.values, cv=skfold)
print("3-Fold Stratified K-Fold Cross Validation Score: \n{}".format(scores3_skfold))
```
