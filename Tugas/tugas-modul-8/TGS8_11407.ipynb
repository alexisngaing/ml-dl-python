```python
# Ubah data string menjadi data numeric

df_stroke["gender"].replace({"Male": 1, "Female": 0, "Other": -1}, inplace=True)
df_stroke["ever_married"].replace({"Yes": 1, "No": 0}, inplace=True)
df_stroke["work_type"].replace({'Private': 0, "Self-employed": 1, "Govt_job": 2, "children": 3, "Never_worked": 4}, inplace=True)
df_stroke["Residence_type"].replace({"Urban": 1, "Rural": 0}, inplace=True)
df_stroke["smoking_status"].replace({"smokes": 2, "formerly smoked": 1, "never smoked": 0, "Unknown": -1}, inplace=True)
```

```python
# Train-test split 

from sklearn.model_selection import train_test_split

X = df_stroke.drop('stroke',axis=1)
y = df_stroke['stroke']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=7
    )

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

cat_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']

transformer = make_column_transformer(
    (OneHotEncoder(), cat_cols),
    remainder='passthrough'
)

X_train_enc = transformer.fit_transform(X_train)
X_test_enc = transformer.transform(X_test)

df_train_enc = pd.DataFrame(X_train_enc,columns=transformer.get_feature_names_out())
df_test_enc = pd.DataFrame(X_test_enc,columns=transformer.get_feature_names_out())

df_train_enc.head(10)
df_test_enc.head(10)
```

```python
# Data Scaling

from sklearn.preprocessing import MinMaxScaler

mm_scaler = MinMaxScaler()

mm_scaler.fit(X_train_enc)
X_train_mmscaled = mm_scaler.transform(X_train_enc)
X_test_mmscaled = mm_scaler.transform(X_test_enc)

print("Minimum before scaling:\n", X_train_enc.min(axis=0))
print("Maximum before scaling:\n", X_train_enc.max(axis=0))
print("Minimum after scaling:\n", X_train_mmscaled.min(axis=0))
print("Maximum after scaling:\n", X_train_mmscaled.max(axis=0))
```

```python
# Grid Search

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X_trainval, X_test, y_trainval, y_test = train_test_split(
    X.values, y.values, test_size=0.3, random_state=42, shuffle=True
)

print("Size of training set:", X_trainval.shape)
print("Size of test set:", X_test.shape)
```

```python
best_score = 0

for g in [0.001,0.01,0.1,1.,10,100]:
    for c in [0.001,0.01,0.1,1.,10,100]:
        svc1 = SVC(gamma=g,C=c)
        svc1.fit(X_trainval,y_trainval)
        
        score = svc1.score(X_test,y_test)
        
        if score > best_score:
            best_score = score
            best_params = {'C':c,'gamma':g}
            
print("Best score: {:.2f}".format(best_score))
print("Best params: {}".format(best_params))
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval, test_size = 0.3, random_state=42, shuffle=True
)

print("Size of training set:", X_train.shape)
print("Size of validation set:", X_val.shape)
print("Size of test set:",X_test.shape)
```

```python
best_score_val = 0

for g in [0.001,0.01,0.1,1.,10,100]:
    for c in [0.001,0.01,0.1,1.,10,100]:
        svc2 = SVC(gamma=g,C=c)
        svc2.fit(X_train,y_train)
        
        score_val = svc2.score(X_val,y_val)
        
        if score_val > best_score_val:
            best_score_val = score_val
            best_params_val = {'C':c,'gamma':g}

svc2 = SVC(**best_params_val)
svc2.fit(X_trainval,y_trainval)
test_score = svc2.score(X_test,y_test)

print("Best score validation: {:.2f}".format(best_score_val))
print("Best params val: {}".format(best_params_val))
print("Test set score with best params: {:.2f}".format(test_score))
```
